Task 02: Data Cleaning and Preparation

Step 1: Import Pandas library

```python
import pandas as pd
```

Step 2: Load the dataset

```python
df = pd.read_csv("/content/samplesuperstore.csv")
```

Step 3: Display first 5 rows

```python
print("First 5 rows:")
print(df.head())
```

Step 4: Check dataset information

```python
print(df.info())
```

Step 5: Identify missing values

```python
print("Missing values:")
print(df.isnull().sum())
```

Step 6: Handle missing numerical values

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

Step 7: Handle missing text values

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Step 8: Remove duplicate rows

```python
df.drop_duplicates(inplace=True)
```

Step 9: Convert string dates to datetime format

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

Step 10: Save cleaned data into a new CSV file

```python
df.to_csv("Cleaned_Global_Superstore.csv", index=False)
```

Step 11: Display success message

```python
print("Cleaning completed successfully")
```
